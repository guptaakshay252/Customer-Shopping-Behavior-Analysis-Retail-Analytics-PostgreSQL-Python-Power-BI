# Read the file

```python
import pandas as pd

file_path = r"C:\Users\LENOVO\Documents\Project\customer_shopping_behavior.csv"

df = pd.read_csv(file_path)

df.head()

```

# Dataset Overview

```python
print("Shape (rows, columns):", df.shape)
print("\nColumn names:\n", df.columns.tolist())

print("\nInfo:")
print(df.info())

print("\nFirst 5 rows:")
display(df.head())

print("\nLast 5 rows:")
display(df.tail())
```

# Check Missing and duplicate values

```python
print("\nMissing values per column:")
missing = df.isnull().sum()
print(missing)

print("\nMissing values (%) per column:")
print((missing / len(df) * 100).round(2))

print("\nTotal missing values:", missing.sum())

print("\nNumber of duplicate rows:", df.duplicated().sum())
```

```python
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
df.isnull().sum()
```

# Summary Statistics

```python
# Numeric columns summary
print("\nNumeric columns summary:")
display(df.describe().T)

# Include object columns too
print("\nSummary including categorical columns:")
display(df.describe(include='all').T)
```

# Checking Unique Values

```python
print("\nNumber of unique values per column:")
print(df.nunique())

# For important categorical columns (adjust names as per your file)
cat_cols = ['Customer_ID', 'Gender', 'Category', 'Payment_Method', 'Location']  # change based on actual columns

for col in cat_cols:
    if col in df.columns:
        print(f"\nValue counts for {col}:")
        print(df[col].value_counts().head(10))
```

# Renaming columns according to snake casing for better readability and documentation


```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df=df.rename(columns={"purchase_amt_(usd)" : "purchase_amt"})
df.columns
```

# Create New Column Age_Grp and purchase_frequency_days



```python
labels=["Young Adult","Adult","Middle_aged","Senior"]
df["age_grp"]=pd.qcut(df["age"],q=4,labels=labels)
print(df[["age","age_grp"]].head(10))

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
print(df[['purchase_frequency_days','frequency_of_purchases']].head(10))
```
